```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
```

```python
working_dir = ''
# working_dir = '/content/drive/MyDrive/Colab Notebooks/k-digital/[Project]Display_Sensor_Anomaly_Analysis/'
fact_data = pd.read_csv(working_dir + 'fact_data_rlc_columns.csv', index_col='dat...date.name.')
```

## VIF계수를 구하는 함수
- 매개변수 : 데이터프레임, 컬럼명 리스트
- 반환값 : VIF 데이터프레임, 가장큰 VIF를 가진 컬럼명, 가장 큰 VIF계수

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(df, columns_all):
#     print('\n',columns_all)
    # 매개변수로 받은 각 피처마다의 VIF 계수를 출력
    vif_df = pd.DataFrame()
    vif_df["VIF Factor"] = [variance_inflation_factor(df[columns_all].values, i) for i in range(df[columns_all].values.shape[1])]
#     print(df[columns_all].columns)
    vif_df["features"] = columns_all
    vif_df.sort_values(by="VIF Factor",ascending=False, inplace=True)
#     print(vif_df)    
    # 가장 큰 vif 컬럼명
    max_col_name = vif_df.iloc[0].features
    # 가장 큰 vif계수
    max_vif = max(vif_df['VIF Factor'])
    
    return vif_df, max_col_name, max_vif
```

## 382개의 모든 컬럼(RIGHT, LEFT, CENTER 포함)에 대한 VIF계수를 구하고 컬럼 삭제
- 컬럼수 382개 => 179 개로 감소
- 총 3시간 30분 소요

```python
column_list = list(fact_data.columns)

new_vif_df = pd.DataFrame()
new_max_col_name = ''
new_max_vif = 100
count = 0
while new_max_vif > 30:
    count += 1
    new_vif_df, new_max_col_name, new_max_vif = get_vif(fact_data, column_list)
    column_list.remove(new_max_col_name)
    print(f'[{count}] 삭제된 컬럼명 : ', new_max_col_name)
    print(f'[{count}] VIF계수 : ', new_max_vif)
    print('\n')
    
new_vif_df
```

# 데이터 프레임 저장
- new_vif_df.csv : VIF계수가 30이하로 줄어든 VIF데이터프레임 저장
- fact_data_vif.csv : 줄어든 컬럼을 적용한 데이터프레임 저장

```python
new_vif_df.to_csv(working_dir + 'new_vif_df.csv')
fact_data_vif = fact_data[column_list]
fact_data_vif.to_csv(working_dir + 'fact_data_vif.csv')
```
